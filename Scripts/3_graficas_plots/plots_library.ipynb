```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear la figura y los subgráficos
fig, axes = plt.subplots(nrows=3, #Numero de filas
                         ncols=1, #Numero de columnas
                         figsize=(8, 6))# Tamaño de la imagen ( cm )

# Primera gráfica
# como vas a graficar y que vas  agraficar 
axes[0].plot(x,# datos en eje X
            y1,# atos en columna Y
            color='b', #color
            label='Seno') #Etiqueta 
axes[0].set_title('Función Seno')
axes[0].legend()
axes[0].grid()

# Segunda gráfica
axes[1].plot(x, y2, color='r', label='Coseno')
axes[1].set_title('Función Coseno')
axes[1].legend()
axes[1].grid()

# Ajustar el espacio entre las gráficas
plt.tight_layout()

# Mostrar la figura
plt.show()
```
